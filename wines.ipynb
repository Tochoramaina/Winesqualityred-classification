# OUTLIER ANALYSIS

```python
# a boxplot to show aoutliers in chlorides
sns.boxplot(x=df['chlorides'])
plt.title('outlier detection')
```

chlorides have outliers which increase noise in the data thus we are going to remove or replace them

```python
# statistics
df.chlorides.describe()
```

below are quantiles that are used to clean data by  replacing values that are above 95% percentile by replacing them with upper quntile of 75%.

```python
# quantiles necessary to clean data
print(df['chlorides'].quantile(0.25))
print(df['chlorides'].quantile(0.75))
print(df['chlorides'].quantile(0.95))
```

```python
# 0.09 represents the upper quantile values that is used to replace values that are above 95% percentile values
df['chlorides'] = np.where(df['chlorides'] > 0.126099999, 0.09,  df['chlorides'])
```

```python
# visualising chlorides that have fewer outliers
sns.boxplot(x=df['chlorides'])
plt.title('cleaned chlorides boxplot')
```

The outliers are left so thus the data is not overfitted. Some noise is required for an improved performance of the model

```python
sns.boxplot(x=df['total_sulfur_dioxide'])
plt.title('Outlier detection in total sulfur dioxide')
```

```python
# A function to replace/renove outliers
def replace_outlier(df, x):
    print(df[x].quantile(0.25))
    print(df[x].quantile(0.75))
    print(df[x].quantile(0.95))
    i = df[x].quantile(0.95)
    y = df[x].quantile(0.75)
    df[x] = np.where(df[x] > i, y, df[x])
    sns.boxplot(x=df[x])
    plt.title(f'Outlier cleaning in {x}')
    print(f'The outliers of {x} have been modified')
    return df
```

```python
# replacing outliers of total sulfur dioxide
df = replace_outlier(df, 'total_sulfur_dioxide')
```

```python
# quality and their frequency
df.quality.value_counts()
```

```python
# cleaning quality by replacing values that are below 5 by 0, and above 5 by 1
df['quality'] = np.where(df['quality'] > 5, 1, 0)
```

```python
# splitting the dataset into x and y variables
x = df.drop('quality', axis=1)
y = df['quality']
```

```python
# x and y variables are then splitted into training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=40)
```

standardscaler is also called variace scaling since it normalises values into scaled features of between 0 and 1.It uses this formulae of x - np.mean(x)/np.sqrt(np.var(x)) to feature scale data.

```python
# applying standard scaler to normalise data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# modelling 
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
# fitting our model
lr.fit(x_train, y_train)
accuracy = lr.score(x_test,y_test)
print(accuracy)
```

Logistic regression has an accuracy of 78%

```python
# applying RandomforestClssifier model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=20)
# fitting our model
clf.fit(x_train, y_train)
acc = clf.score(x_test,y_test)
pred = clf.predict(x_test)
print(acc)
```

RandomForestClasssifier has an accuracy of 86%

```python
# 
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
print(classification_report(pred, y_test))
```

```python
cm = confusion_matrix(pred,y_test, labels=clf.classes_)
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= clf.classes_)
display.plot()
plt.title('Confusion Matrix')
```
